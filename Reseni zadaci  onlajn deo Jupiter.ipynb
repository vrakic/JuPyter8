**Задатак 1.** _Индекс телесне тежине_ (енгл. _body mass index_, BMI) представља количник телесне масе изражене у килограмима и квадрата телесне висине у изражене у метрима:

$$\hbox{BMI} = \frac{\hbox{masa u kilogramima}}{(\hbox{visina u metrima})^2}$$

У наредној ћелији дати су подаци о тежини (у килограмима) и дужини/висини (у центиметрима) једног дечака од његовог шестог месеца па до његових 6,5 година.

```python
merenja   = ["6 мес", "1,5 год", "2,5 год", "3,5 год", "4,5 год", "5,5 год", "6,5 год"]
masaKG    = [5.9,     11.5,      14.8,      20.5,      22.0,      24.2,      29.0     ]
visinaCM  = [62.0,    84.0,      97.0,      115.0,     122.5,     131.5,     135.0    ]
```

Додати нову ћелију испод ове и у њој графиконом приказати промену његовог индекса телесне тежине.

```python
BMI = []
for i in range(7):
    visina_M = visinaCM[i] / 100
    indeks = masaKG[i] / (visina_M * visina_M)
    BMI += [indeks]

import matplotlib.pyplot as plt
plt.plot(merenja, BMI)
plt.show()
plt.close()
```
